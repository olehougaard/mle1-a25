# Preamble

```python
import numpy as np

import matplotlib.pyplot as plt
import math
```

```python
x = (np.arange(0, 200, 2) * .1)
mean = 10
sigma = 4
y = [1 / (2*math.pi*sigma**2) * math.exp(-math.pow(x - mean, 2)/(2 * sigma**2)) for x in x]
plt.plot(x, y)
```

```python
X_prime = 5 * x - 20
plt.plot(X_prime, y)
```

```python
plt.boxplot(x, vert = False)
```

```python
X_exp = [math.pow(1.5, t) for t in x]
plt.plot(X_exp, y[:len(X_exp)])
```

```python
X_sq = [t**2 for t in x]
plt.plot(X_sq, y)
```

```python
plt.boxplot(X_sq, vert = False, showmeans=True)
```

```python
X_sqrt = [math.sqrt(t) for t in x]
plt.plot(X_sqrt, y)
```

```python
plt.boxplot(X_sqrt, vert = False, showmeans=True)
```

```python
X_log = [math.log(t) for t in x if t > 0]
plt.plot(X_log, y[1:])
```

```python
from sklearn.datasets import make_circles
from matplotlib.colors import ListedColormap

colors = ListedColormap(['green', 'magenta', 'yellow', 'blue'])

X_circles, y_circles = make_circles(n_samples=300, factor=.4, noise = .08)
plt.scatter(X_circles[:, 0], X_circles[:, 1], c=y_circles, cmap=colors)
```

```python
def polar(p):
    x, y = p
    r = np.sqrt(x**2 + y**2)
    theta = np.arctan2(x, y)
    return [r, theta]

X_polar = np.array([polar(p) for p in X_circles])
plt.scatter(X_polar[:, 0], X_polar[:, 1], c=y_circles, cmap=colors)
```

```python
minutes = np.arange(0, 72 * 60)
y = [(minute % 1440) / 60 for minute in minutes]
x = minutes / 60
plt.xlabel('time (hours)')
plt.ylabel('time of day (hours)')
plt.plot(x, y)
```

```python
plt.ylabel('sin(time of day)')
rady = [h * np.pi / 12 for h in y]
plt.plot(x, np.sin(rady))
```

```python
plt.ylabel('cos(time of day)')
rady = [h * np.pi / 12 for h in y]
plt.plot(x, np.cos(rady))
```
