```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

from sklearn.datasets import make_blobs

from sklearn.tree import DecisionTreeClassifier
```

```python
blobs = make_blobs(centers=[[-2.5, 0], [2.5, 4], [8.5, -1.5]], cluster_std= 2.5, n_samples=300, random_state=504)

X, y = blobs

classes = ["green", "yellow", "magenta"]

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=ListedColormap(classes))
```

```python

tree_clf = DecisionTreeClassifier(random_state=504, min_impurity_decrease=.02)
tree_clf.fit(X, y)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

y_hat = tree_clf.predict(X)

cm = confusion_matrix(y, y_hat)
accuracy = accuracy_score(y, y_hat)
precision = precision_score(y, y_hat, average=None)
recall = recall_score(y, y_hat, average=None)
f1 = f1_score(y, y_hat, average=None)

print("Confusion Matrix:\n", cm, "\n")

print("Accuracy: ", accuracy)

print("\tPrec\tRec\tf1-score")
for i in range(len(classes)):
    print("{}\t{:.2f}\t{:.2f}\t{:.2f}".format(classes[i], precision[i], recall[i], f1[i]))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

dis = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)
dis.plot()
```
