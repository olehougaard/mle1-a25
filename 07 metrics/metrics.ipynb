```python
import numpy as np
from sklearn.metrics import ConfusionMatrixDisplay

cm = np.array([[999996, 0], [4, 0]])

dis = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
dis.plot()
```

```python
cm = np.array([[0, 999996], [0, 4]])

dis = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
dis.plot()
```

```python
cm = np.array([[999976, 20], [1, 3]])

dis = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
dis.plot()
```

```python
TN = cm[0][0]
FP = cm[0][1]
FN = cm[1][0]
TP = cm[1][1]

sum = TN + TP + FN + FP
positives = TP + FP
actuals = TP + FN
acc = (TN + TP) / sum
prec = TP / positives
rec = TP / actuals
f1 = 2 * prec * rec / (prec + rec)

print("accuracy = ", acc)
print("precision = ", prec)
print("recall = ", rec)
print("f1-score = ", f1)
```
